# Stops Words and Stop Words Removal

````python
from nltk.corpus import stopwords 
````

````python
stop_words = stopwords.words('english')
````

````python
stop_words[0:5]
````

````python
tokens[0:10]
````

````python
new_tokens = [word for word in tokens if word.lower() not in stop_words]
````

````python
new_tokens = []
for word in tokens:
    if word.lower() not in stop_words:
        new_tokens.append(word)
````
